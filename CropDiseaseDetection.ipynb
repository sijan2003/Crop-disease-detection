```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths
base_dir = '/root/.cache/kagglehub/datasets/emmarex/plantdisease/versions/1/PlantVillage'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')

# Create directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Split images for each class
for class_name in os.listdir(base_dir):
    class_path = os.path.join(base_dir, class_name)
    if os.path.isdir(class_path) and class_name not in ['train', 'validation']:  # Avoid nested directories
        # Create subdirectories in train and validation directories
        train_class_path = os.path.join(train_dir, class_name)
        val_class_path = os.path.join(val_dir, class_name)
        os.makedirs(train_class_path, exist_ok=True)
        os.makedirs(val_class_path, exist_ok=True)

        # Get all images (exclude directories)
        images = [img for img in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, img))]

        # Create labels (class for each image)
        labels = [class_name] * len(images)  # All images in this class have the same label (class name)

        # Split into train and validation sets with stratification based on labels
        train_images, val_images = train_test_split(images, test_size=0.2, stratify=labels, random_state=42)

        # Move images to respective directories
        for img_name in train_images:
            shutil.copy(os.path.join(class_path, img_name), os.path.join(train_class_path, img_name))
        for img_name in val_images:
            shutil.copy(os.path.join(class_path, img_name), os.path.join(val_class_path, img_name))

print("Data successfully split into training and validation sets!")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set up ImageDataGenerator for loading and augmenting training images
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Rescaling pixel values to between 0 and 1
    rotation_range=40,  # Randomly rotate images to avoid overfitting
    width_shift_range=0.2,  # Shift the image horizontally
    height_shift_range=0.2,  # Shift the image vertically
    shear_range=0.2,  # Shear the image
    zoom_range=0.2,  # Random zoom
    horizontal_flip=True,  # Flip images randomly
    fill_mode='nearest'  # Fill empty areas after transformations
)

# Set up ImageDataGenerator for validation images (no augmentation)
val_datagen = ImageDataGenerator(rescale=1./255)

# Load the images from directories into the generators
train_generator = train_datagen.flow_from_directory(
    train_dir,  # Path to training directory
    target_size=(150, 150),  # Resize all images to 150x150 pixels
    batch_size=32,  # Load images in batches of 32
    class_mode='categorical',  # Use categorical labels (one-hot encoded)
    shuffle=True  # Shuffle training data for better generalization
)

val_generator = val_datagen.flow_from_directory(
    val_dir,  # Path to validation directory
    target_size=(150, 150),  # Resize all images to 150x150 pixels
    batch_size=32,  # Load images in batches of 32
    class_mode='categorical',  # Use categorical labels (one-hot encoded)
    shuffle=False  # No need to shuffle validation data
)
```

```python
from tensorflow.keras import layers, models

# Define the CNN model
model = models.Sequential([
    # Convolutional layer with 32 filters and 3x3 kernel size, followed by ReLU activation
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D((2, 2)),  # Max pooling layer to reduce dimensions

    layers.Conv2D(64, (3, 3), activation='relu'),  # Convolutional layer with 64 filters
    layers.MaxPooling2D((2, 2)),  # Max pooling layer

    layers.Conv2D(128, (3, 3), activation='relu'),  # Convolutional layer with 128 filters
    layers.MaxPooling2D((2, 2)),  # Max pooling layer

    layers.Flatten(),  # Flatten the 3D output to 1D
    layers.Dense(512, activation='relu'),  # Fully connected layer with 512 units
    layers.Dense(train_generator.num_classes, activation='softmax')  # Output layer for multi-class classification
])

# Compile the model with the Adam optimizer, categorical cross-entropy loss function, and accuracy metric
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary to check architecture details
model.summary()
```

```python
# Importing EarlyStopping from Keras
from tensorflow.keras.callbacks import EarlyStopping

# Train the model using the training and validation generators
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    train_generator,  # Training data generator
    steps_per_epoch=train_generator.samples // train_generator.batch_size,  # Number of steps per epoch
    epochs=10,  # Number of epochs to train the model
    validation_data=val_generator,  # Validation data generator
    validation_steps=val_generator.samples // val_generator.batch_size,  # Number of validation steps
    callbacks=[early_stopping]  # Use early stopping to prevent overfitting
)

# Print a message after training is complete
print("Model training completed!")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import numpy as np

# Evaluate the model and get predictions
val_loss, val_acc = model.evaluate(val_generator, steps=val_generator.samples // val_generator.batch_size)
y_pred_probs = model.predict(val_generator)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true = val_generator.classes


# Accuracy and Loss Line Plot (if history is available)
plt.figure(figsize=(14, 6))

try:
    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

except AttributeError:
    print("No training history available to plot accuracy/loss curves.")
```

```python
from PIL import Image
import numpy as np

def preprocess_image_pillow(img_path):
    # Load the image using Pillow
    img = Image.open(img_path)

    # Resize the image to (150, 150)
    img = img.resize((150, 150))

    # Convert the image to a numpy array
    img_array = np.array(img)

    # Add batch dimension (because the model expects a batch of images)
    img_array = np.expand_dims(img_array, axis=0)

    # Scale the image values to (0, 1)
    img_array = img_array / 255.0

    return img_array
```
